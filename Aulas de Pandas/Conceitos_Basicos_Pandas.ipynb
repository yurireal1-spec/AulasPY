## 4. Implementar a solução passo a passo

### 4.1 Criar um DataFrame

Um DataFrame é uma estrutura de dados tabular, similar a uma planilha.

Pseudocódigo:
- Importar dados de dicionário ou lista
- Criar DataFrame com pd.DataFrame()

```python
# Criar um DataFrame de exemplo
data = {
    'Nome': ['Ana', 'Bruno', 'Carlos', 'Diana'],
    'Idade': [25, 30, 35, 28],
    'Cidade': ['São Paulo', 'Rio', 'Belo Horizonte', 'Porto Alegre'],
    'Salário': [5000, 6000, 5500, 5200]
}
df = pd.DataFrame(data)
print("DataFrame criado:")
df
```

### 4.2 Selecionar e filtrar dados

Selecionar colunas ou linhas específicas.

Pseudocódigo:
- Usar df['coluna'] para coluna
- Usar df.loc[] ou df.iloc[] para linhas
- Filtrar com condições: df[df['coluna'] > valor]

```python
# Selecionar coluna
print("Coluna Nome:")
print(df['Nome'])

# Filtrar pessoas com idade > 28
filtro = df[df['Idade'] > 28]
print("\nPessoas com idade > 28:")
filtro
```

### 4.3 Limpar dados

Lidar com valores ausentes, tipos de dados incorretos.

Pseudocódigo:
- Verificar missing: df.isnull()
- Remover: df.dropna()
- Preencher: df.fillna()

```python
# Adicionar valores ausentes para exemplo
df.loc[1, 'Salário'] = np.nan
print("DataFrame com missing:")
print(df)

# Verificar missing
print("\nValores ausentes:")
print(df.isnull().sum())

# Preencher missing com média
df['Salário'] = df['Salário'].fillna(df['Salário'].mean())
print("\nApós preencher:")
df
```

### 4.4 Agregar dados

Usar groupby para agrupar e calcular estatísticas.

Pseudocódigo:
- Agrupar: df.groupby('coluna')
- Agregar: .sum(), .mean(), etc.

```python
# Adicionar coluna para agrupamento
df['Região'] = ['Sul', 'Sudeste', 'Sudeste', 'Sul']

# Agrupar por região e calcular média do salário
agregado = df.groupby('Região')['Salário'].mean()
print("Média salarial por região:")
print(agregado)
```

## 5. Visualizar dados e resultados

Criar gráficos para ilustrar os dados e validar o raciocínio.

```python
# Gráfico de barras para salários
plt.figure(figsize=(8, 5))
plt.bar(df['Nome'], df['Salário'], color='skyblue')
plt.title('Salários por Pessoa')
plt.xlabel('Nome')
plt.ylabel('Salário')
plt.show()

# Scatter plot: Idade vs Salário
plt.figure(figsize=(8, 5))
plt.scatter(df['Idade'], df['Salário'], color='red')
plt.title('Idade vs Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.grid(True)
plt.show()
```

## 6. Comparar abordagens e discutir resultados

Comparar diferentes formas de manipular dados no Pandas, como usar .loc vs .iloc, ou diferentes métodos de agregação.

Discussão: Pandas oferece flexibilidade; escolher o método certo depende do contexto para eficiência.

```python
# Comparar seleção com .loc e .iloc
print("Seleção com .loc (rótulo):")
print(df.loc[0:2, 'Nome':'Idade'])

print("\nSeleção com .iloc (posição):")
print(df.iloc[0:3, 0:2])

# Discussão: .loc usa rótulos, .iloc usa índices numéricos. Escolher baseado na necessidade.
```
